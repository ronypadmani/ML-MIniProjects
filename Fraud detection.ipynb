Part 1: Overview of Dataset

```python
# import necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df=pd.read_csv('creditcard.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
#size of the dataset
df.shape
```

```python
# to check the data types of each column and non null values
df.info()
```

```python
# all null value
df.columns
```

```python
#shows the five number summary and mean and standard deviation for each column
df.describe()
```

```python
df['Class'].value_counts()
```

```python
df['Class'].value_counts().rename( index={ 0:'Not fraud', 1:'Fraud'})
```

```python
#plot the pie chart for visually show the distribution of data

values=df['Class'].value_counts()
labels=['Not fraud','Fraud']

plt.figure(figsize=(10,6))
plt.pie(values, labels=labels, autopct='%1.2f%%',startangle=90)
plt.show()
```

Here we see that our dataset is highly imbalanced but we not require to modify it because always the no of not fraud is very high compare to no of fraud one.
Here we detect the 0.17% of fraud transaction.

```python
# seperating the target column and other columns :
X=df.drop(columns=['Class'])
X
```

```python
# define target variable Y for classification
Y=df['Class']
Y
```

```python
# divide our dataset into two part train and test with test size is = 25%
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split( X,Y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sd=StandardScaler()

X_train_sc=sd.fit_transform(X_train)
X_test_sc=sd.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(X_train_sc, y_train)
```

```python
y_pred_lr=lr.predict(X_test_sc)
```

```python
# check the accuracy
from sklearn.metrics import *

print('Accuracy of Logistic Regression model',accuracy_score(y_test,y_pred_lr)*100)
```

Here we get a 99.9% accuaracy this is because of our imbalanced dataset , so for better and efficient evaluation we use confusion matrix..

```python
# confusion matrix
c_mat_lr=confusion_matrix(y_test,y_pred_lr)

sns.heatmap(c_mat_lr, annot=True,cmap='coolwarm')
plt.title('Confusion Matrix for Logistic Regression :')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

By showing this confusion matrix we say that in 44 transaction model predict the Not fraud while in actual case it is fruad so we need to reduce this number in any way

So we use different model and check their evaluation or confusion matrix

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()

dt.fit(X_train_sc, y_train)
```

```python
y_pred_dt=dt.predict(X_test_sc)
```

```python
print('Accuracy of Decision Tree Model :',accuracy_score(y_test, y_pred_dt)*100)
```

```python
# confusion matrix
c_mat=confusion_matrix(y_test,y_pred_dt)

sns.heatmap(c_mat, annot=True,cmap='coolwarm')
plt.title('Confusion Matrix For Decision Tree :')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

**we see that here the value is reduced 44 to 25 means our Decision tree model works better than Logistic Regression**
